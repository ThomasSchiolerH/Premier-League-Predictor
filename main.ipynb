## Merge datasets

```python
# Combine the team names from both datasets to ensure consistent encoding
all_teams = pd.concat([results['home_team'], results['away_team'], stats['team']]).unique()

# Create a mapping for team names to unique codes
team_code_mapping = {team: code for code, team in enumerate(all_teams)}

# Apply the encoding to the results dataset
results['home_code'] = results['home_team'].map(team_code_mapping)
results['away_code'] = results['away_team'].map(team_code_mapping)

# Apply the encoding to the stats dataset
stats['team_code'] = stats['team'].map(team_code_mapping)

# Display the mapping and check the results
team_code_mapping, results.head(), stats.head()
```

```python
# Merge home team stats
results_with_home_stats = results.merge(stats, left_on=['home_code', 'season'], right_on=['team_code', 'season'], how='left', suffixes=('', '_home'))

# Merge away team stats
results_with_full_stats = results_with_home_stats.merge(stats, left_on=['away_code', 'season'], right_on=['team_code', 'season'], how='left', suffixes=('', '_away'))

# Drop the redundant team_code columns from the stats after merging
results_with_full_stats = results_with_full_stats.drop(columns=['team_code', 'team_code_away'])

# Display the first few rows to check the merged data
results_with_full_stats.head()
```

```python
# Export the merged dataset to a CSV file
results_with_full_stats.to_csv('merged_results_with_stats.csv', index=False)

```

## Missing data

```python
# Check for missing values in the dataset
missing_values = results_with_full_stats.isnull().sum()

# Display columns with missing values
missing_values[missing_values > 0]
```

#### Treating missing values

```python
# Fill missing values with 0 for any missing data
# Separate numerical and categorical columns
numerical_cols = results_with_full_stats.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = results_with_full_stats.select_dtypes(include=['object', 'category']).columns

# Fill numerical columns with mean
results_with_full_stats[numerical_cols] = results_with_full_stats[numerical_cols].fillna(results_with_full_stats[numerical_cols].mean())

# Fill categorical columns with mode
results_with_full_stats[categorical_cols] = results_with_full_stats[categorical_cols].fillna(results_with_full_stats[categorical_cols].mode().iloc[0])

```

### Feature scaling, categorial

```python
# Encode the 'result' column (target variable) into numeric values
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
results_with_full_stats['result_encoded'] = label_encoder.fit_transform(results_with_full_stats['result'])

# Define the target variable
y = results_with_full_stats['result_encoded']


```

### Split data into trainging and test sets

```python
# Ensure 'season' is in datetime format
results_with_full_stats['season'] = pd.to_datetime(results_with_full_stats['season'].str.split('-').str[0], format='%Y')

# Sort data by season
results_with_full_stats = results_with_full_stats.sort_values('season')

# Use seasons before the last one for training and the last season for testing
train_data = results_with_full_stats[results_with_full_stats['season'] < results_with_full_stats['season'].max()]
test_data = results_with_full_stats[results_with_full_stats['season'] == results_with_full_stats['season'].max()]

# Prepare features and target
features_to_drop = ['result', 'result_encoded', 'home_team', 'away_team', 'season']

X_train = train_data.drop(columns=features_to_drop)
y_train = train_data['result_encoded']
X_test = test_data.drop(columns=features_to_drop)
y_test = test_data['result_encoded']


# Encode if any categorical columns in X_train
non_numeric_columns = X_train.select_dtypes(include=['object', 'category']).columns

# Combine X_train and X_test for consistent encoding
for column in non_numeric_columns:
    combined_data = pd.concat([X_train[column], X_test[column]], axis=0)
    combined_data_encoded = combined_data.astype('category').cat.codes
    X_train[column] = combined_data_encoded.iloc[:X_train.shape[0]].values
    X_test[column] = combined_data_encoded.iloc[X_train.shape[0]:].values



# Check for missing values in X_train and X_test
print("Missing values in X_train:", X_train.isnull().sum().sum())
print("Missing values in X_test:", X_test.isnull().sum().sum())

```

### Training the model

```python
# Initialize the RandomForest model
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=42)

# Train the model
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)
```

```python
y_train.value_counts()
```

### Accuracy

```python
# Evaluate the model
from sklearn.metrics import accuracy_score, classification_report

accuracy = accuracy_score(y_test, y_pred)
print(f"Model accuracy: {accuracy:.4f}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=label_encoder.classes_))
```

### Confusion matrix

```python
combined = pd.DataFrame(dict(actual=y_test, predicted=y_pred))
conf_matrix = pd.crosstab(index=combined["actual"], columns=combined["predicted"])
print(conf_matrix)
```
